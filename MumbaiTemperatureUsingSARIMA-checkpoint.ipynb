```python
import pandas as pd
import numpy as np
from numpy import log
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
from datetime import timedelta
from statsmodels.tsa.statespace.sarimax import SARIMAX

temp_df = pd.read_excel("E:\Coursera\TabPy\Github\MumbaiTemperature.xlsx",engine='openpyxl')
mumbai_df = temp_df[['dt','AverageTemperature']]
mumbai_df['dt'] =  pd.to_datetime(mumbai_df['dt'], format="%Y%m%D")
mumbai_df.set_index('dt',inplace = True)

plt.figure(figsize=(12,6))
plt.plot(mumbai_df['AverageTemperature'])
```

```python
mumbai_df['month'] = mumbai_df.index.month
mumbai_df['year'] = mumbai_df.index.year
year_avg = pd.pivot_table(mumbai_df, values='AverageTemperature', index='year', aggfunc='mean')
year_avg['5 Years MA'] = year_avg['AverageTemperature'].rolling(5).mean()
year_avg[['AverageTemperature','5 Years MA']].plot(figsize=(12,6))
plt.title('Yearly AVG Temperatures in Mumbai')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.xticks([x for x in range(1980,2010,3)])
plt.show()
```

## There seems to be a slight upward trend in Mean over the years.

# First Differencing

```python
mumbai_firstdiff = mumbai_df.diff().dropna()
mumbai_firstdiff['month'] = mumbai_firstdiff.index.month
mumbai_firstdiff['year'] = mumbai_firstdiff.index.year

new_avg = pd.pivot_table(mumbai_firstdiff, values='AverageTemperature', index='year', aggfunc='mean')
new_avg['5 Years MA'] = new_avg['AverageTemperature'].rolling(5).mean()
new_avg[['AverageTemperature','5 Years MA']].plot(figsize=(12,6))
plt.title('Yearly AVG Temperatures in Mumbai')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.xticks([x for x in range(1980,2010,3)])
plt.show()
```

## Mean is now centered around zero

# Plot ACF and PACF

```python
y = plot_acf(mumbai_firstdiff['AverageTemperature'], lags=36)
x = plot_pacf(mumbai_firstdiff['AverageTemperature'], lags=36)
```

## ACF - Lag 12 seems to be the most significant lag

## PACF - 1st lag is positive and the next positive lag is 12. 

### These suggest seasonal AR & MA Process

# ADF Statistics

```python
def check_stationarity(y):
    adfresultset = adfuller(y,autolag='AIC')
    adfresult = pd.Series(adfresultset[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
    adfresult = round(adfresult,4)
    
    for key, value in adfresultset[4].items():
        adfresult["Critical Value (%s)"%key] = value.round(4)
        
    print(adfresult)

check_stationarity(mumbai_firstdiff['AverageTemperature'])
```

# SARIMA Model Summary

```python
train_end = datetime(2008,12,1)
test_end = datetime(2010,12,1)

train_data = mumbai_firstdiff[:train_end]['AverageTemperature']
test_data = mumbai_firstdiff[train_end + timedelta(days=1):test_end]['AverageTemperature']

pdq = (0,1,0)
PDQM = (1, 0, 1, 12)

model = SARIMAX(train_data, order=pdq, seasonal_order=PDQM)
model_fit = model.fit()
print(model_fit.summary())
```

# Forecast and Residuals

```python
predictions = model_fit.forecast(len(test_data))
predictions = pd.Series(predictions, index=test_data.index)
residuals = test_data - predictions
```

```python
plt.figure(figsize=(10,4))
plt.plot(residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Residuals from SARIMA Model', fontsize=20)
plt.ylabel('Error', fontsize=16)
```

# Data Vs Predictions

```python
plt.figure(figsize=(20,8))

plt.plot(mumbai_firstdiff['AverageTemperature'])
plt.plot(predictions)

plt.legend(('Data', 'Predictions'), fontsize=16)
```

```python
print('Mean Absolute Percent Error:', round(np.mean(abs(residuals/test_data)),4))
```

```python
print('Root Mean Squared Error:', np.sqrt(np.mean(residuals**2)))
```
